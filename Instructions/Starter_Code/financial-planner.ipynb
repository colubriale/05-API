### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret

alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

# Create the Alpaca API object
api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

```python
# Format current date as ISO format
# Set start and end datetimes 
start_date = pd.Timestamp("2017-08-22", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-08-07", tz="America/New_York").isoformat()

# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
# By setting the start date to "2017-08-20" and the enddate to "2021-08-07" we will get 252 days per year and using the 1000 limit 
# which is the limit for free Alpaca data we will be able to get almost 4 years of data

df_ticker = api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,
).df

# Preview DataFrame
# df_ticker.head(5) 
df_ticker
```

```python
# Get the last row as this would be the lates info on the shares
df_ticker_row = df_ticker[-2:-1]
agg_close_price = float(df_ticker_row["AGG"]["close"])
df_ticker_row 
```

```python

agg_close_price = float(df_ticker_row["AGG"]["close"])
spy_close_price = float(df_ticker_row["SPY"]["close"])

# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
my_agg_value = my_agg * agg_close_price
my_spy_value = my_spy * spy_close_price

# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
monthly_income =12000

# Consolidate financial assets data
amounts1 = [{"amount": my_btc_value + my_eth_value },
            {"amount": my_agg_value + my_spy_value }]

# Create savings DataFrame
df_savings = pd.DataFrame(amounts1, index =["crypto","shares"])  

# Display savings DataFrame
#display(df_savings)
print(df_savings)                                          
```

```python
# Plot savings pie chart
df_savings.plot.pie(y="amount", title="My Crypto and Share Portfolio")
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
# I could have used my_btc_value + my_eth_value + my_agg_value + my_spy_value but the following is shorter
total_amount_savings = df_savings["amount"].sum()

# Validate saving health

# I use the following to test the logic
# total_amount_savings = emergency_fund - 200

if total_amount_savings > emergency_fund :
    print(f"Congtulations you have enough savings for an emergency fund")
elif total_amount_savings == emergency_fund :
    print(f"Congtulations you have exactly the amount for an emergency fund")
else :
    print(f"Sorry you do not have enough savings for an emergency fund. You need ${emergency_fund - total_amount_savings} more") 
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
# Format current date as ISO format
# the dates were chosen based ipon the Alpaca Free version max of 1000 records. We will be able to get almost 4 years of data

start_date = pd.Timestamp("2017-08-22", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-08-07", tz="America/New_York").isoformat()
```

```python
# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

df_stock_data = api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,
).df


# Display sample data
df_stock_data.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
# Typical trading year is 252 days
# We have 4 years ie 252 * 4 days of past trading data
# We want 30 years simulation ie 252 * 30 days 
# We also have a 60% of AGG and 40% of SPY so we use that fraction ratio for weight

MC_30_years = MCSimulation(
    portfolio_data = df_stock_data,
    weights = [.60,.40],
    num_simulation = 500,
    num_trading_days = 252 * 30
)
```

```python
# Printing the simulation input data
df_stock_data
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
MC_30_years.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_30_years.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_30_years.plot_distribution()
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
MC_30_years_stats = MC_30_years.summarize_cumulative_return()

# Print summary statistics
print(MC_30_years_stats)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
ci_lower = round(MC_30_years_stats[8] * initial_investment,2)
ci_upper = round(MC_30_years_stats[9] * initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
ci_lower = round(MC_30_years_stats[8] * initial_investment,2)
ci_upper = round(MC_30_years_stats[9] * initial_investment,2)

# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
